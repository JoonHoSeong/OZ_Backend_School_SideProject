```python
from sqlalchemy import create_engine
from sqlalchemy import MetaData
from faker import Faker
from faker_airtravel import AirTravelProvider
from tools.control_table import *
from tools.dummy_generator import *
from tqdm import tqdm

engine = create_engine(f'mysql+pymysql://root:root@localhost:3306/airportdb?charset=utf8mb4', echo=True)
metadata = MetaData()
metadata.reflect(engine)
# Faker 객체 초기화
seed_num = random.randint(0,9999)
Faker.seed(seed_num)
faker = Faker()
faker.add_provider(AirTravelProvider)
```

```python
def run_create_dummy(select_table, fake_data_num, method, engine, metadata) :
    if  method :
        truncate_table(selected_table=select_table, engine=engine)
    if select_table == 'airline' :
        data = create_airline_dummy_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'airport' :
        data = create_airport_dummy_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'airplane_type' :
        data = create_airplanetype_dummy_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'airplane' :
        data = create_airplane_dummy_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'airport_geo' :
        data = create_airportgeo_dummy_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'airport_reachable' :
        data = create_airport_reachable_dummy_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'employee' :
        data = create_employee_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'flight_log' :
        data = create_flight_log_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'flight' :
        data = create_flight_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'flightschedule' :
        data = create_flightschedule_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'passenger' :
        data = create_passenger_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'weatherdata' :
        data = create_weatherdata_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    elif select_table == 'passengerdetails' :
        data = create_passengerdetails_data(faker= faker, fake_data_num=fake_data_num)
        insert_dummy_data(selected_table=select_table, engine=engine, metadata=metadata, data=data)
    
```
